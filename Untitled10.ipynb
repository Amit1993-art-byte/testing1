```python
#Q.5 Define ndarrays in NumPy and explain their key features. How do they differ from standard python lists .


 #ANS: ndarray (n-dimensional array) in NumPy is a multidimensional container of items of the same type and size. It is essentially

 #    a grid of values all of same type , indexed by a tuple of non negative integers.

  #   The number of dimension is called the rank.

   #  key features and Difference from standard python lists :

  #   1. Homogeneous Data types

   #    All elements in an ndarray have the same data type ,unlike python lists which can have elements of different data types.

   #    this allows for more efficient and faster computation.

  #   2. Multidimesional support

   #    Unlike python lists ndarray supports multiple dimensions 1D, 2D , 3D etc .


  #   3. Vectorized opeartion

    #   NumPy allows to perform operations on entire arrays without the need for loops . This is called vectorization and it increases

   #    speed of computation.


  #  4. Indexing and Slicing

   #   Like python lists , ndarrays can be sliced and indexed but with more advanced options like multi-dimensional indexing and slicing.

  #  5. Broadcasting

   #   It performs operations between arrays of differnt shapes whereas python does not support braodcasting .










```

```python
# Q.6 Analyse the performance benefits of NumPy arrays over python lists for large scale numercial  operations .

#  ANS: NumPy arrys have significant advantages over python lists .

 #     1. Memory efficiency

 #      NumPy arrays store in contiguous blocks of memory ,which leads to faster access and improved performance .

 #    2. vectorization

  #     It apply operations on entire arrays at once without writing explicit loops in python .


  #   3. Element wise operations

   #    Addition substarction etc are much faster in NumPy arrays than in python lists because they are implemeted in

   #    low-level languages like c or fortran.


   #  4. paralleslism

   #    Many NumPy operations are internally parallelized taking advnatge of multiple cores or CPU's which significantly improves performance.
```
