```python
#Q. 5 what are iterators in python and how they are different from iterables?

# ANS: Iterators in python are an object that represents a stream of data. They are used to iterate over a sequence of values, one at a time.

 #     They works by implementing the iterator protocol, which consists of the __iter__() and __next__() methods.

 #     Example: my_list= [1,2,3]
  #             iterator =iter(my_list)

  #             print(next(iterator)) #output:1
  #             print(next(iterator)) #output:2
  #             print(next(iterator)) #output:3

   #   Iterables are objects that can be iterated over, meaning they can be used in a for loop.

   #   Example: my_list = [1,2,3]

   #            for item in my_list:
   #                print(item)

   #                output: 1
   #                        2
   #                        3

    #     key difference between itearator and  iterables

    #     1. memory usage

    #        iterators are memory efficient as they compute the next element only when it is requested whereas iterables can hold all elements in

     #       memory (ex: list, tuples etc)


    #      2. reusability

     #        iterators are not reusable but iterables are reusable as many times we want.


```

```python
#Q 7. what are the advantage of using generators over regular functions?

# ANS: The advantage of using generators over regular functions is that they allow us to generate a sequence of values on the fly,

 #     without having to store them all in memory at once.

 #     1. Memory efficiency

  #       Generators are memory efficient as they generate values one at a time, and only when they are requested.

  #    2. Lazy evaluation

   #      generators yield values on demand so they can be used to work with potentially infinite sequences, like streams of data without running out of memory.


  #    3. simplified code

   #      more concise and readable code by eliminating   the need to manage the state of the iteration explicitly.


```
